```python
import os
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
os.environ["OPENAI_MODEL_NAME"] = "gpt-3.5-turbo"
```

```python
import crewai
from crewai_tools import SerperDevTool
```

```python
serper_tool = SerperDevTool(api_key="d210f6f367df4602520fca04df19e62bc9426767")
```

```python
research_agent = crewai.Agent(
    role="Web Researcher",
    goal="Search the internet and summarize information from credible sources.",
    backstory="You are a skilled research analyst who finds and summarizes the latest information using online sources.",
    verbose=True,
    allow_delegation=False,
    tools=[serper_tool],
)
```

```python
def search(query_or_topic):
  research_task = crewai.Task(
    description= f"Find and summarize on the query or topic: {query_or_topic}",
    expected_output="A clear and concise 2-3 paragraph summary of required topic/query.",
    agent=research_agent,
    verbose=True
  )
  crew = crewai.Crew(
    agents=[research_agent],
    tasks=[research_task],
    verbose=True
  )
  result=crew.kickoff()

  return result
```

```python
result = search("What is the lastest invention or research on Quantum Computing?")
```

```python
with open("SerperOutput.txt", "w") as f:
  f.writelines(result.raw)
```
