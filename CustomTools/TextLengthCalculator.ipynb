```python
from typing import Type
from crewai.tools import BaseTool
from crewai import Agent, Crew, Task
from pydantic import BaseModel, Field
```

```python
class MyToolInput(BaseModel):
  text: str = Field(..., description="The text to count words in")
```

```python
class WordCountTool(BaseTool):
  name: str = "word_count"
  description: str = "Counts the number of words in a given text"
  args_schema: Type[BaseModel] = MyToolInput

  def _run(self, text: str):
    length = len(text)
    return f"The given text has {length} characters."
```

```python
length_tool = WordCountTool()
```

```python
agent = Agent(
    role="Text Inspector",
    goal="Analyze text for its properties.",
    backstory="You are a linguistics expert who measures and interprets text metrics.",
    verbose=True,
    allow_delegation=False,
    tools=[length_tool]
)
```

```python
sentence = 'Yes â€” when building custom tools for CrewAI, you can (and should, for more complex tools) use Pydanticâ€™s BaseModel and the Type hinting pattern.'
task = Task(
    description=f"Find out how many characters are in this sentence: {sentence}",
    expected_output="Return the character count.",
    verbose=True,
    agent=agent
)
```

```python
crew = Crew(
    agents=[agent],
    tasks=[task],
    verbose=True
)
```

```python
result = crew.kickoff()
```
